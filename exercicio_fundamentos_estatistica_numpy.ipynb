### **1.1. Agregação** 

Neste primeira atividade, irei gerar um array NumPy por dia. Para cada array irei somar todos os incidentes que aconteceram naquela meia hora. Sendo assim, cada array deve ter 27 posições, cada qual com a soma dos incidentes daquela meia hora.

**Dica**: remover a primeira e a última coluna.

```python
df.shape
```

```python
import numpy as np
import pandas as pd

df = pd.read_csv('traffic.csv', sep=';')

data14 = np.array(df.iloc[0:27,1:-1].sum(axis=1))
data15 = np.array(df.iloc[27:54,1:-1].sum(axis=1))
data16 = np.array(df.iloc[54:81,1:-1].sum(axis=1))
data17 = np.array(df.iloc[81:108,1:-1].sum(axis=1))
data18 = np.array(df.iloc[108:135,1:-1].sum(axis=1))

print('14:', data14)
print('15:', data15)
print('16:', data16)
print('17:', data17)
print('18:', data18)
```

### **1.2. Métricas** 

Para cada array você deve calcular as seguintes métricas: 

 - **média**;
 - **desvio padrão**.

```python
media14 = np.mean(data14)
media15 = np.mean(data15)
media16 = np.mean(data16)
media17 = np.mean(data17)
media18 = np.mean(data18)

print('Média de 14:', media14)
print('Média de 15:', media15)
print('Média de 16:', media16)
print('Média de 17:', media17)
print('Média de 18:', media18)
```

```python
desvio_padrao14 = np.std(data14)
desvio_padrao15 = np.std(data15)
desvio_padrao16 = np.std(data16)
desvio_padrao17 = np.std(data17)
desvio_padrao18 = np.std(data18)

print('Desvio padrão de 14:', desvio_padrao14)
print('Desvio padrão de 15:', desvio_padrao15)
print('Desvio padrão de 16:', desvio_padrao16)
print('Desvio padrão de 17:', desvio_padrao17)
print('Desvio padrão de 18:', desvio_padrao18)
```

### **1.3. Interpretação** 

Baseado nos resultados da questão 1.2, responda:

> Qual dia apresenta a maior média de acidentes por meia hora?

**Resposta:** Dia 16
